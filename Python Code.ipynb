The next idea was to explore the impact of China's Cultural Revolution on development indicators. Does the data agree with the sentiments I hear from my elders who lived through it, specifically that life was "miserable"? I have my reservations about the quality or accuracy of data from China, especially during that time period, but I am hoping the overall patterns will still be truthful.

```python
china = df[df['Country Name'] == 'China'].reset_index(drop=True).drop('Country Name', axis=1)
china = china.set_index('Indicator Name').T
china.rename_axis('Year', inplace=True)

# Assign era (pre, during, post Cultural Revolution)
era = china.index.to_list()
for i in era:
    if int(i) < 1966:
        era[era.index(i)] = 'Pre-Cultural Revolution'
    elif int(i) <= 1976:
        era[era.index(i)] = 'Cultural Revolution'
    else:
        era[era.index(i)] = 'Post-Cultural Revolution'        
china.insert(loc=0, column='Era', value=era)

china.describe()
```

The first thing I noticed about China's data is the large amount of NaNs especially before the 2000s. This is unsurprising given how relatively undeveloped China was at the time, as well as some potential political reasons surrounding government transparency or ability to collect data that is too complex for the scope of this project. In any case, I decided to remove any rows (indicator) that contained NaN(s) for any year in order to perform PCA, unsurprisingly shortening the list of indicators from 1516 all the way down to 111.

```python
chinaDropNA = china.dropna(axis=1, how='any')
pd.DataFrame(chinaDropNA.columns.to_list()).to_csv('chinaDropNAIndicators.csv')
chinaDropNA.head()
```

I found that there weren't enough relevant indicators. As such, I narrowed the data to only be between 1960 to 1982. The Cultural Revolution took place from 1966-1976, meaning the earliest date of 1960 is 6 years before the event. Hence I picked 1982 for symmetry as the end date of our data. Removing any rows with NaN for only 1960-1982 resulted in 39 more indicators being kept, including intuitively important ones such as mortality rate and life expetancy.

```python
china1982 = china.loc['1960':'1982']
china1982 = china1982.dropna(axis=1, how='any')
pd.DataFrame(china1982.columns.to_list()).to_csv('china1982Indicators.csv')
china1982.head(5)
```

I started with PCA to see if any indicators would stand out (i.e. if the first few principal components would contain high absolute loading value with specific original variables/indicators). If so, I could narrow down to these indicators in a time series graph to see how the Cultural Revolution affected them.


```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

dat = StandardScaler().fit_transform(china1982.drop('Era', axis=1)) # Standardize values

# pca1982 = PCA(n_components=china1982.shape[0]) # observation count < variable count so take the observation count amount of PCs
pca = PCA(n_components=None) 

components = pca.fit_transform(dat)
```

```python
labels = {}
totalVariance = 0
for i in range(len(pca.explained_variance_ratio_)):
    if totalVariance >= 0.89:
        break
    labels[str(i)] = f"PC {i+1} ({pca.explained_variance_ratio_[i]*100:.1f}%)"
    totalVariance += pca.explained_variance_ratio_[i]

fig = px.scatter_matrix(
    components,
    labels=labels,
    dimensions=range(len(labels)),
    color=china1982["Era"],
    title=f'Total Explained Variance: {totalVariance*100:.2f}%'
)
fig.update_traces(diagonal_visible=False)
fig.update_layout(
    font=dict(color='black', size=16),
    autosize=False,
    width=1500,
    height=1000,
    legend=dict(
        orientation="h",  # Horizontal orientation
        yanchor="bottom", # Anchor the bottom of the legend to the y position
        y=1.02,           # Position slightly above the plot area (normalized coordinates)
        xanchor="right",  # Anchor the right of the legend to the x position
        x=1,              # Position at the right edge
        font=dict(color='black', size=16)
    ),
    legend_title_text=''
)

fig.show()
```

Particularly with the first 2 principal components, it is clear that data values do vary based on the Cultural Revolution. However, it is difficult to say in which ways, as it is improbable to plot all the loadings (original indicators) onto the PC plot. I wanted to see if I could only plot the most significant ones, but realized quickly that there was no meaningful cutoff for the number

```python
loadings = pd.DataFrame(pca.components_, columns=china1982.drop('Era', axis=1).columns)

plt.hist(loadings.iloc[0].abs(), edgecolor='black')
plt.title('Absolute Loadings of PC 1')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

Furthermore, it is entirely possible that the variance in data is simply due to time/growth. After all, the three eras in the PCA plots are not clearly seperated (each group is followed by another in space). Whether the Cultural Revolution contradicted a general trend in the indicators over time or not would best be observed using a time series graph.
As such, I moved on from PCA to find other methods of selecting indicators for "quality of life." 

At first I thought to regain some data lost while removing indicators that had NaNs. After all, that was for PCA, but without PCA, it was potentially acceptable to have some missing values for some years. Of course, it would be meaningless to include indicators that have only missing values, or even indicators that consisted primarily of missing values, especially during the main time period of study, 1960-1982. Hence, I dropped indicators that had missing values for more than 1/3 of the years from 1960-1982. This resulted in 183 remaining indicators.

```python
china1982TS = china.loc['1960':'1982']

# Calculate the threshold for columns (2/3 the number of rows)
thresh_cols = int(np.ceil(len(china1982TS) * 2 / 3))

# Drop columns where less than 'thresh_cols' non-NA values are present
china1982TS = china1982TS.dropna(axis=1, thresh=thresh_cols)

pd.DataFrame(china1982TS.columns.to_list()).to_csv('china1982TSIndicators.csv')
china1982TS.shape
```

I spent quite some time deciding how to choose indicators from the 183. There were 2 main issues: 
1. There is no objective "best list" of indicators for the quality of life for a country. While many resources online may point to general indicators, such as wealth, life expectancy, education, unemployment, access to healthcare or electricity, air and water quality, etc., the WDI dataset contains too many detailed indicators. If my approach was to choose indicators in a vacuum first, then look for any of them in the 183 indicators, no matter what the decisions would be somewhat arbitrary.
2. It would be time-consuming to research relevant indicators, only to find them missing in the 183 (i.e. Gini index).

Hence, I ultimately opted for a more straightforward approach: manual selection. While it certainly feels like an unelegant and non-scalable method, it saved me time from the second point above. And to the first point, there will always be subjectivity in what constitutes "the best indicators" for quality of life. My belief is that I followed the most efficient method given the data: picking from 183 indicators after dropping the ones that had too many missing values based on intuition and how they matched with online resources regarding quality of life indicators. 

```python
TSIndicators = pd.read_csv('TS_Indicators.csv')
chinaTS = china[TSIndicators['Indicator Name'].to_list()]
chinaTS = chinaTS.rename(columns={'Life expectancy at birth, total (years)': 'Life expectancy at birth (years)', 
                                  'Population, total': 'Population',
                                  'Age dependency ratio (% of working-age population)': 'Age dependency ratio (%)'})
era = chinaTS.index.to_list()
for i in era:
    if int(i) < 1966:
        era[era.index(i)] = 'Pre-Cultural Revolution'
    elif int(i) <= 1976:
        era[era.index(i)] = 'Cultural Revolution'
    else:
        era[era.index(i)] = 'Post-Cultural Revolution'        
chinaTS.insert(loc=0, column='Era', value=era)
chinaTS
```

```python
corr = chinaTS.corr(numeric_only=True)
corr

# Correlation heat plot
plt.figure(figsize=(10,10))

sns.heatmap(corr,
            center=0,
            linewidths=.1,
            annot=True,
            linecolor ='black',
            cmap = 'RdBu',
            square=True)

plt.title('Correlation Heatmap of Selected "Quality of Life" Indicators for China', fontsize=12, fontweight='bold')
plt.legend()
```

```python
chinaTS_normalized = chinaTS.copy(deep=True).drop('Era', axis=1)
chinaTS_normalized[chinaTS_normalized.columns] = StandardScaler().fit_transform(chinaTS_normalized[chinaTS_normalized.columns])

chinaTS_normalized.rename(columns={'Life expectancy at birth (years)': 'Life Expectancy (z-score)',
                                 'Population': 'Population (z-score)',
                                 'Physicians (per 1,000 people)': 'Physicians (z-score)',
                                 'Food production index (2014-2016 = 100)': 'Food Production Index (z-score)',
                                 'GDP per capita (constant 2015 US$)': 'GDP per Capita (z-score)'}, inplace=True)

plt.figure(figsize=(10, 6))
for ind in chinaTS_normalized.columns:
    plt.plot(chinaTS_normalized.index, chinaTS_normalized[ind], linestyle='-', label=ind)

plt.xticks(np.arange(0, 70, 10))
plt.axvspan(6, 16, facecolor='lightcoral', alpha=0.2, label='Cultural Revolution')
plt.legend()
plt.show()
```
